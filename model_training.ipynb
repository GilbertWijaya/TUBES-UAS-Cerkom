```python
from tensorflow.keras.models import Sequential # untuk mendefinisikan layer pada cnn/ann
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense # layer pada cnn untuk pengolahan dan manipulasi data
from tensorflow.keras.preprocessing.image import ImageDataGenerator # untuk mendefinisikan konfigurasi data training gambar
from tensorflow.keras.optimizers import Adam # untuk optimisasi pelatihan model
```

```python
# Dataset generator
train_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2) # menyiapkan dataset disertai konfigurasi preprocessing awal
# rescale = melakukan pemetaan nilai piksel berikisar 0 - 1
```

```python
# preprocessing image
train_data = train_gen.flow_from_directory(
    'dataset',
    target_size=(64, 64),
    batch_size=8,
    class_mode='binary',
    subset='training'
)
val_data = train_gen.flow_from_directory(
    'dataset',
    target_size=(64, 64),
    batch_size=8,
    class_mode='binary',
    subset='validation'
)
```

```python
# Model CNN ringan
model = Sequential([
    Conv2D(16, (3,3), activation='relu', input_shape=(64,64,3)),
    MaxPooling2D(2,2),
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Latih model (cepat, hanya 10 epoch)
model.fit(train_data, validation_data=val_data, epochs=10)
```

```python
# Simpan model
model.save('model_egg.h5')
print("Model disimpan ke model_egg.h5")
```
